```python
W = tf.Variable(tf.zeros([784,10]))
with tf.Session() as sess:
    sess.run(W.initializer)
    print(sess.run(W))
```

```python
W = tf.Variable(tf.zeros([784,10]))
with tf.Session() as sess:
    sess.run(W.initializer)
    print(W.eval())
```

```python
W = tf.Variable(tf.zeros([784,10]))
W_assign = W.assign(tf.fill([784, 10], 9.0))
with tf.Session() as sess:
    sess.run(W_assign)
    print(W.eval())
```

```python
x1 = tf.Variable(9)
sess1 = tf.Session()
sess2 = tf.Session()
x1_assign1 = x1.assign_add(3)
x1_assign2 = x1.assign_sub(5)

sess1.run(x1.initializer)
sess2.run(x1.initializer)

print(sess1.run(x1_assign1))
print(sess2.run(x1_assign2))

sess1.close()
sess2.close()
```

```python
W = tf.Variable(tf.truncated_normal([700, 10]))
U = tf.Variable(W * 2)
with tf.Session() as sess:
    sess.run(W.initializer)
    sess.run(U.initializer)
    print(sess.run(W))
    print(sess.run(U))
```

```python
sess = tf.InteractiveSession()
a = tf.constant(3)
b = tf.constant(5)
c = tf.add(a, b)
print(c.eval())
sess.close()
```

```python
x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')
z = tf.add(x, y)

with tf.Session() as sess:
	sess.run(tf.global_variables_initializer())
	writer = tf.summary.FileWriter('./graphs/l2', sess.graph)
	for _ in range(10):
		sess.run(z)
	print(tf.get_default_graph().as_graph_def())
	writer.close()
```

```python
x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')

with tf.Session() as sess:
	sess.run(tf.global_variables_initializer())
	writer = tf.summary.FileWriter('./graphs/l2', sess.graph)
	for _ in range(10):
		sess.run(tf.add(x, y))
	print(tf.get_default_graph().as_graph_def()) 
	writer.close()
```
